```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['runs'] = df['runs'].astype(int)  # Convert 'runs' column to integer type if it's not already
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed_cumulative'] = df.groupby('match_id')['player_dismissed'].cumsum()

df['wickets_left'] = 10 - df['player_dismissed_cumulative']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# Convert 'runs' column to numeric, handling errors by converting non-numeric values to NaN
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Drop rows with missing values in the 'runs' column (optional)
# df.dropna(subset=['runs'], inplace=True)

# Calculate rolling sum for each group
rolling_sum_per_group = df.groupby('match_id')['runs'].rolling(window=30).sum()

# Extract rolling sum values for each group
last_five = rolling_sum_per_group.reset_index(drop=True).tolist()
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Define preprocessing steps for numeric and categorical features
numeric_features = ['current_score', 'balls_left', 'wickets_left', 'crr', 'last_five']
categorical_features = ['batting_team', 'bowling_team', 'city']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler(with_mean=False))  # Specify with_mean=False
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps for numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# Define the model
model = LogisticRegression()

# Create the pipeline with preprocessing and model steps
pipe = Pipeline(steps=[('preprocessor', preprocessor), ('logistic_regression', model)])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Make predictions on the test data
y_pred = pipe.predict(X_test)

# Evaluate the model
print("R-squared score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```
